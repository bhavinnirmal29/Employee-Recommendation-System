```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import re
from IPython.display import Image
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("Book1.csv")
```

```python
# display the first rows in the data
data.head()
```

```python
# check for null values
data.isnull().sum()
```

```python
data.drop(data.columns[data.columns.str.contains('unnamed',case = False)],axis = 1, inplace = True)
```

```python
data.head()
```

```python
f,ax=plt.subplots(1,2,figsize=(12,6))
data[' Placed'].value_counts().plot.pie(explode=[0,0.1],autopct='%1.1f%%',ax=ax[0],shadow=True)
ax[0].set_title(' Placed')
ax[0].set_ylabel('')
sns.countplot(' Placed',data=data,ax=ax[1])
ax[1].set_title('Placed')
plt.show()
```

```python
data.isnull().sum()
```

```python
data = pd.get_dummies(data)
data.head()
```

```python
# Preparing training data and evaluation data
train_labels = data[' Placed'].values[:100]
train_data = data.drop(' Placed', axis=1).values[:100]
eval_labels = data[' Placed'].values[24:]
eval_data = data.drop(' Placed', axis=1).values[24:]
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(fit_intercept=True)
clf.fit(train_data, train_labels)
```

```python
print ('Accuracy of the model on train data: {0}'.format(clf.score(train_data, train_labels)))
```

```python
print ('Accuracy of the model on eval data: {0}'.format(clf.score(eval_data, eval_labels)))
```
